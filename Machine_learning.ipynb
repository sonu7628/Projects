```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target

def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((np.array(point1) - np.array(point2))**2))

# KNN prediction
def knn_predict(training_data, training_labels, test_point, k):
    distances = []
    for i in range(len(training_data)):
        dist = euclidean_distance(test_point, training_data[i])
        distances.append((dist, training_labels[i]))
    distances.sort()
    k_nearest_labels = [label for _, label in distances[:k]]
    
    res = {}
    for label in k_nearest_labels:
        res[label] = res.get(label, 0) + 1

    return max(res, key=res.get)

def accuracy(test_data, test_labels, training_data, training_labels, k):
    correct = 0
    for i in range(len(test_data)):
        pred = knn_predict(training_data, training_labels, test_data[i], k)
        if pred == test_labels[i]:
            correct += 1
    return correct / len(test_data)

training_data = df[["sepal length (cm)", "sepal width (cm)"]].values.tolist()[:120]
training_labels = df["species"].tolist()[:120]
test_data = df[["sepal length (cm)", "sepal width (cm)"]].values.tolist()[120:]
test_labels = df["species"].tolist()[120:]
k = 3
print("Test Accuracy:", accuracy(test_data, test_labels, training_data, training_labels, k))
```
