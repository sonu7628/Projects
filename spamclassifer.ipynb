```python
df['label'] = df['label'].apply(lambda x: 1 if x.strip().lower() == 'spam' else 0)
```

```python
#Converting any text document into lower case
df['message'] = df['message'].str.lower()
```

```python
df
```

```python
#removal of HTML tags
def remove_tags(text):
    pattern=re.compile('<.*>')
    return pattern.sub(r'',text)
```

```python
text="<HTML><body><P>war2"
```

```python
print(remove_tags(text))
```

```python
#remove punctuations
import string
string.punctuation
```

```python
def remove_punctuation(text):
    result = ''
    for char in text:
        if char not in string.punctuation:
            result += char
    return result
df['no_punctuation'] = df['message'].apply(remove_punctuation)
```
